```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
field = "Field_Measurements.csv"
field_df = pd.read_csv(field)

field_df
```

```python
field_df.rename(columns={"T (C)": "Temp (C)"})
```

```python
field_df.count()
```

```python
field_cleanup = field_df [["site number", "site location", "Cluster", "depth (m)", "date", "T (C)", "DO (mg/l)", "pH"]]
field_cleanup = field_cleanup.dropna()
```

```python
field_cleanup.count()
```

```python
field_cleanup
```

```python
temp_x = field_cleanup['pH'];
temp_y = field_cleanup['T (C)']
```

```python
plt.xlabel('pH');
plt.ylabel('Temperature (C)')

plt.scatter (temp_x,temp_y)
```

```python
year_x = field_cleanup ['date'];
year_y = field_cleanup['pH']
```

```python
plt.xlabel('Year')
plt.ylabel('pH')

plt.scatter (year_x, year_y)
```
