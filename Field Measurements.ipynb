This study shows the different pH levels of all the canals and reservoirs that we use here in Phoenix. We are taking data from the following areas:
    AZ Canal, CAP Canal, Chandler, Mesa Turnout, Salt River, Tempe, Verde River
    
pH levels are affected by temperature. The higher the temperature, the pH level drops.

A safe pH level for drinking water is between 6 and 8.5.

High pH water tastes like baking soda and leaves residue on fixtures. The water is also hard. It doesn't cause any health risks, but can be hard to lather with soap, can cause skin problems, and is more difficult to heat up.

Low pH water can have a bitter taste and may cause corrosion on fixtures. It can be unsafe to drink because it can be contaminated with pollutants.

```python
#Dependencies

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Import the CSV file

field = "Field_Measurements.csv"
field_df = pd.read_csv(field)

field_df
```

```python
field_df.count()
```

```python
#Clean up the data

field_cleanup = field_df [["site number", "site location", "Cluster", "depth (m)", "date", "T (C)", "DO (mg/l)", "pH"]]
field_cleanup = field_cleanup.dropna()
```

```python
field_cleanup.count()
```

```python
#Show the new information

field_cleanup
```

```python
#Create a scatter plot showing the correlation of the temperature vs the pH level

temp_x = field_cleanup['pH'];
temp_y = field_cleanup['T (C)']
```

```python
plt.figure(figsize=(20,14))
plt.xlabel('pH');
plt.ylabel('Temperature (C)')

plt.scatter (temp_x,temp_y)
```

This data shows that fortunately, with the rise in temperature, we are not seeing a difference in pH levels. The levels are staying consistent for the most part despite the temperature of the water

```python
#Create a scatter plot that shows the pH levels over time

year_x = field_cleanup ['date'];
year_y = field_cleanup['pH']
```

```python
plt.figure(figsize=(20,20))
plt.xlabel('Year')
plt.ylabel('pH')

plt.scatter (year_x, year_y)
```

Here, we are seeing in recent years that the pH levels stayed consistent between 6 and 8.5, with only a few instances of going a little bit higher. However, in the early 2000's, there were multiple instances where the pH level fell very low.
